Imputation Missing Data


```python
columns_with_missing = df.columns[df.isnull().any()].to_list()
```

```python
df = df[~df[columns_with_missing].isnull().all(axis=1)]
```

```python
df.shape
```

```python
missing_condition = ((df.isnull().sum())/ (df.shape[0])) * 100
missing_condition
```

fill missing data - imputation data

```python
# Deleting columns that more than 49% of data missing [country,size]
df.drop(['county', 'size'], axis=1, inplace=True)
```

```python
# fill missing data in year column with median
df['year'].fillna(df['year'].median(), inplace=True)
```

```python
#because these columns and less than 42% of their data are missing and they are object, the best way to fill the miss data with mod
df['condition'].fillna(df['condition'].mode()[0], inplace=True)
df['cylinders'].fillna(df['cylinders'].mode()[0], inplace=True)
df['drive'].fillna(df['drive'].mode()[0], inplace=True)
df['VIN'].fillna(df['VIN'].mode()[0], inplace=True)
df['model'].fillna(df['model'].mode()[0], inplace=True)
df['transmission'].fillna(df['transmission'].mode()[0], inplace=True)
df['type'].fillna(df['type'].mode()[0], inplace=True)
df['manufacturer'].fillna(df['manufacturer'].mode()[0], inplace=True)
df['paint_color'].fillna(df['paint_color'].mode()[0], inplace=True)
```

```python
# Odometer readings are highly dependent on the vehicle's age (year), make (manufacturer), and model.
# fill missing data in odometer column with median of year, manufacture and model
df['odometer'] = df.groupby(['manufacturer', 'model', 'year'])['odometer'].transform(lambda x: x.fillna(x.median()))
# fill remaining with general median
df['odometer'].fillna(df['odometer'].median(), inplace=True)
```

```python
missing_condition = ((df.isnull().sum())/ (df.shape[0])) * 100
missing_condition
```

Delete rows that conatin at the same rows on missing values of these calumns ('fuel', 'title_status', 'image_url', 'description', 'lat', 'posting_date')


```python
columns_with_missing = df.columns[df.isnull().any()].to_list()
df = df[~df[columns_with_missing].isnull().all(axis=1)]
```

```python
missing_condition = ((df.isnull().sum())/ (df.shape[0])) * 100
missing_condition
```

```python
df.shape
```

```python
df['fuel'].fillna(df['fuel'].mode()[0], inplace=True)
df['title_status'].fillna(df['title_status'].mode()[0], inplace=True)
df['description'].fillna(df['description'].mode()[0], inplace=True)
df['lat'].fillna(df['lat'].mode()[0], inplace=True)
df['long'].fillna(df['long'].mode()[0], inplace=True)
```

```python
missing_condition = ((df.isnull().sum())/ (df.shape[0])) * 100
missing_condition
```
